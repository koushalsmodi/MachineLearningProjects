```python
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.uniform(0, 100, size = (30000, 400))
targets = np.random.random(size=(30000,))

data[0]
```

```python
targets[0]
```

```python
max(targets)
```

```python
test_data = np.random.uniform(0, 120, size = (12000,400))
test_targets = np.random.random(size = (12000,))
```

```python
def get_model():
  model = keras.Sequential([
      layers.Dense(512, activation = "relu"),
      layers.Dense(64, activation = "relu"),
      layers.Dense(1)
  ])

  model.compile(optimizer = "rmsprop",
                loss = "mse",
                metrics = ["mae"])

  return model
```

```python
k = 3
num_validation_samples = len(data) // k
print(len(data))
print(len(data) // k)
validation_scores = []

for i in range(k):
  print(f"Processing fold: #{i}")

  validation_data = data[i*num_validation_samples: (i+1)* num_validation_samples]
  validation_target = targets[i*num_validation_samples: (i+1)* num_validation_samples]

  train_data = np.concatenate([
      data[:i*num_validation_samples],
      data[(i+1) * num_validation_samples: ]])

  train_target = np.concatenate([
      targets[:i*num_validation_samples],
      targets[(i+1) * num_validation_samples: ]])

  model = get_model()
  model.fit(train_data, train_target, epochs = 20, batch_size = 128, verbose = 0)

  validation_score = model.evaluate(validation_data, validation_target)
  validation_scores.append(validation_score)
validation_score = np.average(validation_scores)

model = get_model()
model.fit(data, targets, epochs = 20, batch_size = 128)
test_score = model.evaluate(test_data, test_targets)
```
