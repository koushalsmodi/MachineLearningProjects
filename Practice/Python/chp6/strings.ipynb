```python
# Looping and counting

fruit = 'banana'
count = 0

for letter in fruit:
    if letter == 'a':
        count+=1
        
print(count)
        
    
```

```python
# Looping and counting

fruit = 'stale bananas'
count = 0

for i in range(len(fruit)):
    if fruit[i] == 'a':
        print(i, fruit[i])
        count+=1
print(count)
        
    
```
