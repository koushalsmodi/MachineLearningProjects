```python
# Comparison operators
x = 5

if x == 5:
    print(f'{x} equals to 5')
    
if x > 4:
    print(f'{x} is greater than 4')
    
if x >= 5:
    print(f'{x} is greater than or equal to 5')

if x < 6:
    print(f'{x} is less than 6')
    
if x <= 5:
    print(f'{x} is less than or equal to 5')
    
if x != 6:
    print(f'{x} is not equal to 6')
    
```

```python
for i in range(5):
    print(i)
    if i > 2 :
        print(f'{i} is bigger than 2')
    print('Done with i', i)
print('All done')
```

```python
# Nested conditionals

x = 42

if x > 1:
    print(f'{x} is greater than 1')
    
    if x < 100:
        print(f'{x} is lesser than 100')
print('done')
```

```python
# Two-way Decisions

x = 4

if x > 2:
    print('Bigger')
    
else:
    print('Smaller')
    
print('All done')
```

```python
# Multi-way

x = 5

if x < 2:
    print('smaller')
    
elif x < 10:
    print('medium')

else:
    print('larger')
    
print('all done')
```

```python
# Try-except
astr = 'hello bob'

try:
    istr = int(astr)

except:
    istr = -1
    
print('First', istr)

astr = '123'
try:
    istr = int(astr)

except:
    istr = -1

print('Second', istr)
```

```python
user_inp = input('Enter a number: ')

try:
    ival = int(user_inp)

except:
    ival = -1
    
if ival > 0:
    print('nice work')

else:
    print('not a number')
```
