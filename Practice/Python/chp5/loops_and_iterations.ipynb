```python
n = 5

while n > 0:
    print(n)
    n-=1
print('Blastoff!')
print(n)
```

```python
# breaking out of the loop

while True:
    inp = input('> ')
    if inp == 'done':
        break
    print(inp)
print('Done!')
```

```python
# continue -> go back to the top of the loop

while True:
    inp = input('> ')
    if inp[0] == '#':
        continue
    if inp == 'done':
        break
    print(inp)
print('Done!')
```

```python
# definite loops

for i in [5,4,3,2,1]:
    print(i)

print('Blastoff!')
```

```python
friends = ['Ganpati', 'Shiv', 'Alex', 'Bob', 'Saraswati']

for friend in friends:
    print('happy new year!:', friend)

print('Done!')
```

```python
print('Before')
largest_num = -1
for thing in [9, 41, 55, 64, 12, 3]:
    if thing > largest_num:
        largest_num = thing
        
print('After')
print(f'Largest num: {largest_num}')
```

```python
count = 0
print('Before', count)
for num in [9, 41, 55, 64, 12, 3]:
    count+=1
    print(count, num)
    

print('After', count)
```

```python
sum = 0

print('Before', sum)
for num in [9, 41, 55, 64, 12, 3]:
    sum+=num
    print(num, sum)
    

print('After', sum)
```

```python
sum = 0
count = 0

for num in [9, 41, 12, 3, 74, 15]:
    count+=1
    sum+= num
    
avg = sum / count
print(f'After avergage: {avg:.2f}')
```

```python
# Searching using a boolean variable
found = False
for num in [9, 41, 12, 3, 74, 15]:
    if num == 3:
        found = True
        break
        
print(found)
```
