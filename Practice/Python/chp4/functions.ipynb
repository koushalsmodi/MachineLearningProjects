```python
def greet(user):
    if user == 'Namaste':
        print('Hindi')
    
    elif user == 'Hola':
        print('Spanish')
        
    elif user == 'Hello':
        print('English')
    
    else:
        print('Different language')
        
# lang = input('Greeting: ')
greet('Namaste')
greet('Hola')
greet('Hello')
greet('Grt')
```

```python
def greet(user):
    if user == 'Namaste':
        return "Hindi"
    
    elif user == 'Hola':
        return "Spanish"
        
    elif user == 'Hello':
        return "English"
    
    else:
        print('Different language')
        
# lang = input('Greeting: ')
print(greet('Namaste'), 'Glenn')
print(greet('Hola'), 'Glenn')
print(greet('Hello'), 'Glenn')
print(greet('Grt'), 'Glenn')
```
