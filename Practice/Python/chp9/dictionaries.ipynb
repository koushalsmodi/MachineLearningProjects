```python
cabinet = dict()

cabinet['spring'] = 20
cabinet['summer'] = 12
cabinet['fall'] = 3


cabinet['fall'] = cabinet['fall'] + 4
print(cabinet)
```

```python
for key in cabinet.keys():
    print(key)
```

```python
for value in cabinet.values():
    print(value)
```

```python
for key, value in cabinet.items():
    print(key, value)
```

```python
counts = {}

names = ['Seetha', 'Ramesh', 'Geeta', 'Ramesh', 'Krishna', 'Geeta', 'Ramesh']

for name in names:
    counts[name] = counts.get(name, 0) + 1


sorted_counts = dict(sorted(counts.items(), key = lambda item: item[1], reverse=True))
print(next(iter(sorted_counts.items())))
    
```
