## Tool definition

```python
# An agent is separate than a chatbot because of its:
# 1. Ability to take actions
# 2. Perceive the output of those actions
# 3. React accordingly
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain.tools import tool

@tool
def square_root(x: float) -> float:
    """ Calculate the square root of the number """
    return x**0.5
```

```python
square_root.invoke({"x": 24})
```

```python
@tool
def square(x: float) -> float:
    """ Calculate the square of the number """
    return x**2
```

```python
square.invoke({"x": 2})
```

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage, AIMessage

agent = create_agent(
    model = "gpt-5-nano",
    tools = [square_root, square],
    system_prompt="You are an arithmetic wizard. Use your tools to calculate the square root and square of any numbers"
)


question = HumanMessage(content = "What's the square and square root of 467?")

response = agent.invoke(
    {"messages": [question]}
    )

print(response['messages'][-1].content)
```
